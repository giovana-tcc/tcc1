# Jupyter Notebook para criação da grade de fotos
### Figura da solução proposta para visão geral do conjunto de dados

#### Importação de bibliotecas

```python
from PIL import Image
import matplotlib.pyplot as plt
import glob
```

#### Definição de largura e altura da grade de fotos

```python
qtd_width = 14 #quantidade de fotos na largura da imagem
width = 128*qtd_width+15*(qtd_width-1)
```

```python
qtd_height = 7 #quantidade de fotos na altura da imagem
height = 128*qtd_height+15*(qtd_height-1)
```

#### Criando lista com os nomes das imagens a serem utilizadas na grade de fotos

```python
names_list = glob.glob('*.jpeg')
```

#### Criando a imagem base para a grade de fotos

```python
im = Image.new('RGB',(width,height),(255,255,255))

plt.imshow(im)
```

#### Fazendo alguns testes com uma única imagem

```python
teste = Image.open(names_list[0])

im.paste(teste,(0,0))
im.paste(teste,(143,143))
im.paste(teste,(143,0))

plt.imshow(im)
```

#### Construindo a grade de fotos

```python
for i in range(0,width,143):
    for j in range(0,height,143):
        name = names_list.pop()
        aux = Image.open(name)
        im.paste(aux,(i,j))

plt.imshow(im)
```

#### Mostrando a grade de fotos final e salvando na pasta

```python
im.show()
im.save(str(qtd_width)+'x'+str(qtd_height)+'.jpeg')
```
